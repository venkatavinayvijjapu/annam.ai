```python
import pandas as pd
import json
from urllib.request import urlopen
from urllib.error import URLError, HTTPError
```

# Data Collection:

```python
f_data = pd.DataFrame()
```

```python
questions = []
answers = []
# m -> all districts
# j -> year
# k -> months
for m in range(201, 228):
    for j in range(2009, 2023):
        for k in range(1, 13):
            url = f"https://dackkms.gov.in/Account/API/kKMS_QueryData.aspx?StateCD=02&DistrictCd=0{m}&Month={k}&Year={j}"
            try:
                response = urlopen(url)
            except HTTPError as e:
                print('Error code: ', e.code)
            except URLError as e:
                print('Reason: ', e.reason)
            else:
                data = json.loads(response.read())
            if (data['ResponseCode'] == '1'):
                print(m, j, k)
                for i in range(len(data['data'])):
                #     print("ques: " + data['data'][i]['QueryText'], "ans: " +  data['data'][i]['KccAns'])
                    questions.append(data['data'][i]['QueryText'])
                    answers.append(data['data'][i]['KccAns'])
```

```python
f_data['questions'] = questions
f_data['answers'] = answers
```
